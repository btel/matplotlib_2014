```python
import cartopy
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib
matplotlib.rcParams['savefig.dpi'] = 100

import cartopy.crs as ccrs
import numpy as np
from matplotlib import mlab
from goodies import as_table
```

```python
import netCDF4
dataset = netCDF4.Dataset('../data/avhrr-only-v2.20140101.nc')
```

```python
lon = dataset.variables['lon'][:]
lat = dataset.variables['lat'][:]
temp = dataset.variables['sst'][:].squeeze()
```

```python
ax = plt.axes()
ax.contourf(lon, lat, temp, 100)
```

```python
ax = plt.axes(projection=ccrs.Mercator())
ax.coastlines()
ax.add_feature(cartopy.feature.BORDERS)
ax.set_global()
ax.contourf(lon, lat, temp, 20, transform=ccrs.PlateCarree())
plt.savefig('../figures/cartopy_mercator.svg')
```

```python
data = mlab.csv2rec('../data/ratp_arret_graphique_01.csv', 
                      delimiter='#', 
                      names=['nid', 'lon', 'lat', 'name', 'place','type'])
as_table(data)
```

```python
from cartopy.io.img_tiles import MapQuestOSM
imaging = MapQuestOSM()

ax = plt.axes(projection=imaging.crs)
ax.set_extent((2.1873, 2.5066, 48.7894, 48.9288))
ax.add_image(imaging, 12)
plt.plot(data.lon[:], data.lat[:], '.', transform=ccrs.Geodetic(), ms=1)
```

```python
shapes = cartopy.io.shapereader.Reader('../data/rgp_population')
pop = np.array([r.attributes['POPMUN2011']/r.attributes['AREA'] for r in shapes.records()])
```

```python
gl = cartopy.crs.Globe(ellipse='GRS80')
prj = cartopy.crs.LambertConformal(central_latitude=46.5, 
                                   central_longitude=3.,
                                   secant_latitudes=(49, 44),
                                   false_easting=700000,
                                   false_northing=6600000)
```

```python
from matplotlib import colors, cm

ax = plt.axes(projection=imaging.crs)

ax.set_extent((2.0277,2.6662,48.7200,48.9987))
ax.add_image(imaging, 12)

pop_colors = cm.jet(colors.Normalize()(pop))

for c,g in zip(pop_colors, shapes.geometries()):
    ax.add_geometries(g, prj, facecolor=c, alpha=0.5)
plt.plot(data.lon[:], data.lat[:], '.', color='w', transform=ccrs.Geodetic(), ms=1)
```
